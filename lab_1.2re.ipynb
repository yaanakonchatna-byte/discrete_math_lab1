```python
import matplotlib.pyplot as plt
from venn import venn

# Підбираємо дані, щоб заповнити лише потрібні зони для №18
sets = {
    "A": {1, 2},       # Нижня область (поза B)
    "B": {2, 3, 5},    # Внутрішня порожня область
    "C": {4, 5},       # Верхня область
    "D": {4, 6}        # Верхня область (перетин з C)
}

# Формула, яку ми перевіряємо
A, B, C, D = sets["A"], sets["B"], sets["C"], sets["D"]
E = (A | (C & D)) - B

print(f"Елементи зафарбованої області: {E}")

# Візуалізація
venn(sets)
plt.title("Логічна перевірка №18 (X = (A ∪ (C ∩ D)) \ B)")
plt.show()
```
