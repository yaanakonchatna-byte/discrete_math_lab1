```python
# Завдання для варіанту 16
U = {1, 2, 3, 4, 5, 6, 7, 8, 9}
A = {1, 2, 3, 4, 5, 9}
B = {2, 4, 6, 8}
C = {1, 3, 5, 7}
D = {1, 4, 5, 7, 8, 9}

# Множина, яку потрібно отримати (варіант 16)
target = {1, 4, 8, 5, 2, 9, 3, 6, 7}

# Виведення початкових даних
print(f"Множини: \nA={A}, \nB={B}, \nC={C}, \nD={D}")
print(f"Цільова множина: {target}")

# Перевірка: чи можна отримати target через просте об'єднання?
# Наприклад, спробуємо об'єднати A, B та C
result = A | B | C

print("-" * 30)
print(f"Результат об'єднання (A ∪ B ∪ C): {result}")
print(f"Чи збігається з ціллю? {'Так' if result == target else 'Ні'}")
```
