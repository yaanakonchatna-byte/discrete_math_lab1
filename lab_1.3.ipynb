```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# 1. Вхідні дані згідно з Варіантом 16
n_total = 55         # n
languages = 17       # n1 (Додаток A)
fitness = 21         # n2 (Додаток B)
finance = 25         # n3 (Додаток C)

lang_fit = 6         # n12 (A та B)
lang_fin = 5         # n13 (A та C)
fit_fin = 7          # n23 (B та C)
all_three = 4        # n123 (Всі три)

# 2. Побудова діаграми Венна
plt.figure(figsize=(8, 8))
venn = venn3(subsets=(
    languages - lang_fit - lang_fin + all_three, # Тільки Мови (10)
    fitness - lang_fit - fit_fin + all_three,   # Тільки Фітнес (12)
    lang_fit - all_three,                       # Мови + Фітнес (2)
    finance - lang_fin - fit_fin + all_three,   # Тільки Фінанси (17)
    lang_fin - all_three,                       # Мови + Фінанси (1)
    fit_fin - all_three,                        # Фітнес + Фінанси (3)
    all_three                                   # Усі три (4)
), set_labels=('Мови (A)', 'Фітнес (B)', 'Фінанси (C)'))

plt.title("Розподіл респондентів за мобільними додатками (Варіант 16)")
plt.show()

# 3. Розрахунок принаймні одного та тих, хто не користується
at_least_one = languages + fitness + finance - (lang_fit + lang_fin + fit_fin) + all_three
none = n_total - at_least_one

print(f"Використовують принаймні один додаток: {at_least_one}")
print(f"Не користуються жодним додатком: {none}")
```

```python
import matplotlib.pyplot as plt
from venn import venn
# Імпортуємо venn3 про всяк випадок, як у твоєму прикладі
from matplotlib_venn import venn3

# 1. Дані для 4-х множин (A, B, C, D)
sets = {
    "A": {1, 2, 3, 4, 10, 11, 12},
    "B": {3, 4, 5, 6, 12, 13, 14},
    "C": {6, 7, 8, 9, 10, 13, 15},
    "D": {1, 5, 8, 11, 14, 15, 16}
}

# 2. Обчислення твоєї формули: X = D ∪ (A \ B) ∪ (C \ B)
A, B, C, D = sets["A"], sets["B"], sets["C"], sets["D"]
result_set = D | (A - B) | (C - B)

print(f"Елементи зафарбованої області: {result_set}")

# 3. Побудова діаграми
venn(sets)
plt.title("Результат для формули: $X = D \\cup (A \\setminus B) \\cup (C \\setminus B)$")
plt.show()
```
